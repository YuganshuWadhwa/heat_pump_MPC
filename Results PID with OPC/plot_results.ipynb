```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv("T_9_Unisim_PID.csv")

data = pd.DataFrame(df)


# Create a figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces for I_L1real
fig.add_trace(
    go.Scatter(x=data['Timestamp'], y=data['TC Compressor - PV'], mode='lines', name='TC Compressor - PV', line=dict(color='#d62728')),
    secondary_y=False,
)

# Add traces for U_L1real
fig.add_trace(
    go.Scatter(x=data['Timestamp'], y=data['TC Compressor - OP'], mode='lines', name='TC Compressor - OP', line=dict(color='#1f77b4')),
    secondary_y=True,
)

# Update layout
fig.update_layout(
    title='Unisim Advanced PI  -  PV vs OP',
    xaxis_title='Timestamp',
    legend_title='Signals'
)

# Set y-axes titles
fig.update_yaxes(title_text="TC Compressor - PV", secondary_y=False)
fig.update_yaxes(title_text="TC Compressor - OP", secondary_y=True)

# Show figure
fig.show()
```
